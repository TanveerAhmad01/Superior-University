# Data Pre-Processing

```python
df.info()
```

```python
def fillNaObject(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['street','city','statezip','country']
fillNaObject(columns)
```

```python
def fillNaInt(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

columns = ['bedrooms','bathrooms','floors','yr_built']
fillNaInt(columns)
```

```python
def fillNaFloat(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

columns = ['price','sqft_living','sqft_above','sqft_basement']
fillNaFloat(columns)
```

```python
df.info()
```

```python
def convertFloattoInt(cols):
    for i in cols:
        df[i] = df[i].fillna(0).astype('int64')

columns = ['price','bedrooms','bathrooms','floors','yr_built','sqft_living','waterfront','sqft_lot','view','sqft_above','sqft_basement']
convertFloattoInt(columns)
```

# Model Training using SVC

```python
print("-- Training using SVC on Training Data --")
print("-- Parameters & Values: ", end='')

# model_svc = SVC(gamma='auto', random_state=0)
# model_svc.fit(train_x, np.ravel(train_y))

from sklearn.svm import SVR
model_svr = SVR(kernel='rbf')  # rbf kernel is commonly used for regression
model_svr.fit(train_x, train_y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x_scaled = scaler.fit_transform(train_x)
test_x_scaled = scaler.transform(test_x)

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10], 'epsilon': [0.01, 0.1, 0.2], 'kernel': ['linear', 'rbf']}
grid_search = GridSearchCV(SVR(), param_grid, cv=5)
grid_search.fit(train_x_scaled, train_y)
print("Best Parameters: ", grid_search.best_params_)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x_scaled = scaler.fit_transform(train_x)
test_x_scaled = scaler.transform(test_x)
model_svr.fit(train_x_scaled, train_y)
```

```python
# Load saved Model
model = pickle.load(open('model_svc.pkl', 'rb'))
```

```python
model_predictions = model.predict(test_x)
```

```python
testdata_predict = testData.copy(deep=True)
pd.options.mode.chained_assignment = None

testdata_predict['Prediction'] = model_predictions
```

```python
model_accuracy_score = accuracy_score(testdata_predict['price'], testdata_predict['Prediction'])

print("-- Model Accuracy Score: ", end='')
print(round(model_accuracy_score,3))
```
